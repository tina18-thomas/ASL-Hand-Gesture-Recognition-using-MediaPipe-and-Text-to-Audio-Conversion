```python
import pytesseract
import pandas as pd
import cv2

for i, path in enumerate(image_files):  # process all selected good images
    print(f"\n Processing Image {i+1}: {path}")

    # Step 1: Read and preprocess
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Resize large image for better readability
    h, w = gray.shape
    if h > 1500 or w > 1500:
        gray = cv2.resize(gray, (w // 2, h // 2), interpolation=cv2.INTER_AREA)

    # Apply CLAHE to enhance contrast
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    enhanced = clahe.apply(gray)

    # Adaptive thresholding
    thresh = cv2.adaptiveThreshold(enhanced, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                   cv2.THRESH_BINARY, 11, 2)

    # Morphological operations to clean text
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (1,1))
    processed = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

    # Optional: Denoising
    denoised = cv2.fastNlMeansDenoising(processed, h=30)

    # Step 2: OCR
    ocr_data = pytesseract.image_to_data(denoised, output_type=pytesseract.Output.DATAFRAME)
    ocr_data = ocr_data[ocr_data.text.notnull() & (ocr_data.text.str.strip() != '')]

    # Step 3: Reconstruct lines
    lines = ocr_data.groupby('line_num')['text'].apply(lambda x: ' '.join(x)).tolist()

    # Step 4: Print lines
    print("Reconstructed Lines from OCR:")
    for line in lines:
        print(line)
```
